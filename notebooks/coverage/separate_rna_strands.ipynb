```python
in_dir = '/data/parastou/RNAdeg/results/RipRna/bams/'
```

```python
out_dir = '/data/parastou/RNAdeg/results/RipRna/coverage_bams/'
```

### Step 1 : Separate reverse and forward alignments for each bam file

```python
def strand_separate_bam(in_bam, out_dir):
    
    sample = os.path.basename(in_bam).split('.')[0]
    print(in_bam)
    tmp1 = os.path.join(out_dir, sample + '.tmp1.bam')
    tmp2 = os.path.join(out_dir, sample + '.tmp2.bam')
    out1 = os.path.join(out_dir, sample + '.forward.bam')
    out2 = os.path.join(out_dir, sample + '.reverse.bam')
    
    st = pysam.AlignmentFile(open(in_bam, 'rb'))
    r1 = pysam.AlignmentFile(tmp1, 'wb', template=st)
    r2 = pysam.AlignmentFile(tmp2, 'wb', template=st)

    for r in st.fetch(until_eof=True):
        
        if r.is_reverse:
            r1.write(r)
        elif not r.is_reverse:
            r2.write(r)

    r1.close()
    pysam.sort("-o", out1, tmp1)
    pysam.index(out1)

    r2.close()
    pysam.sort("-o", out2, tmp2)
    pysam.index(out2)

    os.remove(tmp1)
    os.remove(tmp2)
    
    print(out1)
    print(out2)
    
    return out1, out2
```

```python
for file in os.listdir(in_dir):
    
    if file.endswith('.bam'):        
        in_bam = os.path.join(in_dir, file)
        strand_separate_bam(in_bam, out_dir)
```
