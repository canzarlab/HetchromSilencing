```python
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
%matplotlib inline
```

```python
myblue = (0.0 / 255, 153.0 / 255, 255.0 / 255)
myviol = (204.0 / 255, 0.0 / 255, 204.0 / 255)
```

```python
# Legend for boxplots
fig, ax = plt.subplots(figsize=(48,6))

patch2 = mpatches.Patch(color='gray', label='protein coding')
patch3 = mpatches.Patch(color=myviol, label='$\it{dg, dh}$')
patch4 = mpatches.Patch(color='green', label='$\it{tlh, SPAC212.10}$')
patch5 = mpatches.Patch(color=myblue, label='heterochromatic\ngenes')

ax.axis('off')

all_handles = (patch2, patch3, patch4, patch5)

leg = ax.legend(handles=all_handles, ncol=4, handlelength=.5, handleheight=.5, fontsize=70, 
                loc='center', borderpad=1.5)
frame = leg.get_frame()
frame.set_linewidth(0)

ax.add_artist(leg)
plt.savefig(os.path.join(out_dir, 'bp_legend.png'))
plt.savefig(os.path.join(out_dir, 'bp_legend.pdf'), format='pdf', bbox_inches='tight')
plt.show()
```

```python
# Legend for boxplots
fig, ax = plt.subplots(figsize=(48,6))

patch2 = mpatches.Patch(color='gray', label='protein coding')
patch3 = mpatches.Patch(color=myviol, label='$\it{dg, dh}$')
patch4 = mpatches.Patch(color='green', label='$\it{tlh, SPAC212.10}$')
patch5 = mpatches.Patch(color=myblue, label='heterochromatic\ngenes')

ax.axis('off')

all_handles = (patch2, patch3, patch4, patch5)

leg = ax.legend(handles=all_handles, ncol=4, handlelength=.5, handleheight=.5, fontsize=60, 
                loc='center', borderpad=1.5)#, columnspacing=2)
frame = leg.get_frame()
frame.set_linewidth(0)

ax.add_artist(leg)
plt.savefig(os.path.join(out_dir, 'bp_legend_sfont.png'))
plt.savefig(os.path.join(out_dir, 'bp_legend_sfont.pdf'), format='pdf', bbox_inches='tight')
plt.show()
```

```python
# Legend for scatter plots
fig, ax = plt.subplots(figsize=(48,6))
ax.axis('off')

a = range(1)
line1 = Line2D(a, a, linewidth=0, color='gray', marker='o', markerfacecolor='white', markersize=50, 
               markeredgewidth=5)
line2 = Line2D(a, a, color='white', marker='o',markerfacecolor=myviol, markersize=50)
line3 = Line2D(a, a, color='white', marker='o',markersize=50, markerfacecolor='green')
line4 = Line2D(a, a, linewidth=0, color='black', marker='o',markersize=50, markerfacecolor='dimgray', 
               markeredgewidth=5)

leg = ax.legend((line1,line2,line3,line4),('genes','${dg, dh}$', '${tlh, SPAC212.10}$', 'ribosomal proteins'),
                numpoints=1, loc=1, ncol=4, fontsize=70)

frame = leg.get_frame()
frame.set_linewidth(0)

plt.savefig(os.path.join(out_dir, 'scatter_legend.png'))
plt.savefig(os.path.join(out_dir, 'scatter_legend.pdf'), format='pdf', bbox_inches='tight')

plt.show()
```
